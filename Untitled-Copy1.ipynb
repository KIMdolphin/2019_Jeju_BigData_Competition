```python
import csv
import datetime
f = open('train.csv', 'r', encoding = 'ANSI')
distancedict = {}
reader = csv.reader(f)
for line in reader:
    if str(line[7])+'_'+str(line[9]) not in list(distancedict.keys()):
        distancedict[str(line[7])+'_'+str(line[9])] = []
    distancedict[str(line[7])+'_'+str(line[9])].append(int(line[13]))
f.close()
```

```python
import numpy as np
distancedictaver = {}
for i in list(distancedict.keys()):
    listaver = sum(distancedict[i])/len(distancedict[i])
    distancedictaver[i] = int(listaver)
```

```python
f = open('train.csv', 'r', encoding = 'ANSI')
distancedict = {}
reader = csv.reader(f)
for line in reader:
    readerlist = []
    datelist = list(map(int, line[1].split('-'))) 
    readerlist.append(datetime.date(datelist[0], datelist[1], datelist[2]).strftime('%A')) #1 요일
    readerlist.append(line[8]) #2 시간대
    readerlist.append(int(line[13])) #3 시간 
    if str(line[7])+'_'+str(line[9]) not in list(distancedict.keys()):
        distancedict[str(line[7])+'_'+str(line[9])] = []
    distancedict[str(line[7])+'_'+str(line[9])].append(readerlist)
f.close()
```

```python
for o in (distancedict.keys()):
    distancedict[o] = list(set(map(tuple, distancedict[o])))
```

```python
distance_weekin_dict = {}
distance_weekend_dict = {}
for i in distancedict.keys():
    distance_weekin_dict[i] = {}
    distance_weekend_dict[i] = {}

for key in list(distancedict.keys()):
    for l in range(len(distancedict[key])):
        if distancedict[key][l][0] in ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday']:
            if not distancedict[key][l][1] in distance_weekin_dict[key] :
                (distance_weekin_dict[key])[distancedict[key][l][1]] = distancedict[key][l][2] - distancedictaver[key]
            else : 
                (distance_weekin_dict[key])[distancedict[key][l][1]] = int(((distance_weekin_dict[key])[distancedict[key][l][1]] + distancedict[key][l][2] - distancedictaver[key]) / 2)
        else:
            if not distancedict[key][l][1] in distance_weekend_dict[key] :
                (distance_weekend_dict[key])[distancedict[key][l][1]] = distancedict[key][l][2] - distancedictaver[key]
            else : 
                (distance_weekend_dict[key])[distancedict[key][l][1]] = int(((distance_weekend_dict[key])[distancedict[key][l][1]] + distancedict[key][l][2] - distancedictaver[key])/2)
```

```python
f = open('test.csv', 'r', encoding = 'ANSI')
result = []
reader = csv.reader(f)
for line in reader:
    long = str(line[7])+'_'+str(line[9])
    b = line[8]
    datelist = list(map(int, line[1].split('-'))) 
    while(long not in distance_weekin_dict):
        for i in distance_weekin_dict:
            if (i.split('_'))[1] == (long.split('_'))[1]:
                long = (i.split('_'))[0]+'_'+str(line[9])
    if datetime.date(datelist[0], datelist[1], datelist[2]).strftime('%A') in ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday']:
        if b not in list(distance_weekin_dict[long]):
            b = distancedictaver[long]
            result.append(distancedictaver[long]+b)
        else:
            result.append(distancedictaver[long]+distance_weekin_dict[long][b])
        
    else :
        if b not in list(distance_weekend_dict[long]):
            b = distancedictaver[long]
            result.append(distancedictaver[long]+b)
        else:
            result.append(distancedictaver[long]+distance_weekend_dict[long][b])                           
f.close()
```
